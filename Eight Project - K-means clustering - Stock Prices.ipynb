```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pylab as pl
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

data=pd.read_csv(r"C:\Users\malay\Downloads\Sample_Stocks.csv")
X=data[["dividendyield"]]
Y=data[["returns"]]
plt.scatter(X,Y)

Z = np.array(data)
```

```python
kmeans=KMeans(n_clusters=3)
kmeans.fit(Z)

centroids=kmeans.cluster_centers_
labels=kmeans.labels_

print(Z)
print('\n')
print(centroids)
print('\n')
print(labels)
```

```python
for i in range(len(Z)):   
    #print("coordinates:", Z[i], "labels", labels[i])
    plt.scatter(Z[i][0], Z[i][1])
    
plt.scatter(centroids[:,0], centroids[:,1], c="black",marker="x", s=150, linewidth=5, zorder=10)
```
