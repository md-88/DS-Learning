# Data Cleaning:
Findings:
- Normalized-losses column contains missing values represented using "?". Let's replace these values
- Remove some of the columns that arent relevant
- Some of the numeric values are represented as object. Need to convert these to float
- The values of fuel-type(gas/diesel) and bodystyle (convertible, hatchback, sedan) are very random and do not matter much

```python
columns_updated= ["normalized-losses", "num-of-doors", "wheel-base", "length", "width", "height", "curb-weight", "bore", 
"stroke", "compression-ratio", "horsepower", "city-mpg", "highway-mpg", "price"]
cars_updated=cars[columns_updated]
cars_updated=cars_updated.replace("?", np.nan)

cars_updated
```

```python
cars_updated.info()
```

```python
cars_updated["horsepower"]=cars_updated["horsepower"].astype("float")
cars_updated["price"]=cars_updated["price"].astype("float")
cars_updated["normalized-losses"]=cars_updated["normalized-losses"].astype("float")
cars_updated["bore"]=cars_updated["bore"].astype("float")
cars_updated["stroke"]=cars_updated["stroke"].astype("float")
```

```python
# Checkout null values
cars_updated.isnull().sum()
```

```python
# Since we are predicting price, lets drop null values in pricing
cars_updated=cars_updated.dropna(subset=["price"])

# Lets drop 2 null values in num-of-doors, small dataset to loose
cars_updated=cars_updated.dropna(subset=["num-of-doors"])

# Change door sizes to numeric 2 or 4 (the values in series was "two" or "four" accordingly)
cars_updated["num-of-doors"].replace(to_replace="four",value="4",inplace=True)
cars_updated["num-of-doors"].replace("two", "2", inplace=True)
cars_updated["num-of-doors"]=cars_updated["num-of-doors"].astype("int64")

# Lets fill other values by mean
cars_updated=cars_updated.fillna(cars_updated.mean())

# Check no more nulls
cars_updated.isnull().sum()
```

```python
# Of the columns to keep, lets normalize the numeric ones so all values range from 0 to 1, except price
backup=cars_updated["price"]
cars_updated=(cars_updated-cars_updated.min())/(cars_updated.max()-cars_updated.min())
cars_updated["price"]=backup

cars_updated
```

# Univariate k-nearest neighbors models

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

def knn_train_test(training_col, target_col, df):
    knn=KNeighborsRegressor()
    np.random.seed(1)
    
    # Randomize order of rows in data frame.
    shuffled_index = np.random.permutation(df.index)
    random_df = df.reindex(shuffled_index)
    
    # Divide dataset in half for train, test
    midsplit=round(len(random_df)/2)

    # Define train and test dataset
    train_set=df.iloc[:midsplit]
    test_set=df.iloc[midsplit:]
    
    # KNN fit and predict
    knn.fit(train_set[[training_col]], train_set[target_col])
    prediction=knn.predict(test_set[[training_col]])
    
    # Calculate mse and rmse
    mse=mean_squared_error(test_set[[target_col]], prediction)
    rmse=np.sqrt(mse)
    return rmse

rmse_results={}
training_columns=cars_updated.columns.drop("price")

# Train the model, and get RMSE for each of the column values
for value in training_columns:
    rmse_value=knn_train_test(value, "price", cars_updated)
    rmse_results[value]=rmse_value
    
# Convert to series and sort
rmse_results_series=pd.Series(rmse_results)
print(rmse_results_series.sort_values())
```

# With different values of k

```python
def knn_train_test(training_col, target_col, df):
    np.random.seed(1)
    
    # Randomize order of rows in data frame.
    shuffled_index = np.random.permutation(df.index)
    random_df = df.reindex(shuffled_index)
    
    # Divide dataset in half for train, test
    midsplit=round(len(random_df)/2)

    # Define train and test dataset
    train_set=df.iloc[:midsplit]
    test_set=df.iloc[midsplit:]
    
    # Different values of k
    k_values = [1,3,5,7,9]
    k_rmse={}
    
    for options in k_values:
        knn=KNeighborsRegressor(n_neighbors=options)

        # KNN fit and predict
        knn.fit(train_set[[training_col]], train_set[target_col])
        prediction=knn.predict(test_set[[training_col]])
    
        # Calculate mse and rmse
        mse=mean_squared_error(test_set[[target_col]], prediction)
        rmse=np.sqrt(mse)
        k_rmse[options]=rmse
       
    return k_rmse

k_rmse_results={}
training_columns=cars_updated.columns.drop("price")

# Train the model, and get RMSE for each of the column values
for value in training_columns:
    rmse_value=knn_train_test(value, "price", cars_updated)
    k_rmse_results[value]=rmse_value
    
# # Convert to series and sort
# rmse_results_series=pd.Series(rmse_results)
k_rmse_results
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

for k,v in k_rmse_results.items():
    x=list(v.keys())
    y=list(v.values())
    
    plt.plot(x,y)
    plt.xlabel("K Value")
    plt.ylabel("RMSE")   
```

```python
# Compute average RMSE across different `k` values for each feature.
feature_avg_rmse = {}
for k,v in k_rmse_results.items():
    avg_rmse = np.mean(list(v.values()))
    feature_avg_rmse[k] = avg_rmse
series_avg_rmse = pd.Series(feature_avg_rmse)
print(series_avg_rmse.sort_values())
```

# Multivariate Model
Let's modify the knn_train_test() function we wrote in the last step to work with multiple columns.

```python
def knn_train_test(training_cols, target_col, df):
    np.random.seed(1)
    
    # Randomize order of rows in data frame.
    shuffled_index = np.random.permutation(df.index)
    random_df = df.reindex(shuffled_index)
    
    # Divide dataset in half for train, test
    midsplit=round(len(random_df)/2)

    # Define train and test dataset
    train_set=df.iloc[:midsplit]
    test_set=df.iloc[midsplit:]
    
    # Different values of k
    k_values = [5]
    mv_k_rmse={}
    
    for options in k_values:
        knn=KNeighborsRegressor(n_neighbors=options)

        # KNN fit and predict
        knn.fit(train_set[training_cols], train_set[target_col])
        prediction=knn.predict(test_set[training_cols])
    
        # Calculate mse and rmse
        mse=mean_squared_error(test_set[[target_col]], prediction)
        rmse=np.sqrt(mse)
        mv_k_rmse[options]=rmse
       
    return mv_k_rmse

mv_k_rmse_results={}
training_columns=cars_updated.columns.drop("price")

# Pass to function top 2, 3, 4, 5, 6 feature set
best_two_features=["highway-mpg", "horsepower"]
two_best_rmse=knn_train_test(best_two_features, "price", cars_updated)
mv_k_rmse_results["two best features"]=two_best_rmse

best_three_features=["highway-mpg", "horsepower", "city-mpg"]
three_best_rmse=knn_train_test(best_three_features, "price", cars_updated)
mv_k_rmse_results["three best features"]=three_best_rmse

best_four_features=["highway-mpg", "horsepower", "city-mpg", "curb-weight"]
four_best_rmse=knn_train_test(best_four_features, "price", cars_updated)
mv_k_rmse_results["four best features"]=four_best_rmse

best_five_features=["highway-mpg", "horsepower", "city-mpg", "curb-weight", "width"]
five_best_rmse=knn_train_test(best_five_features, "price", cars_updated)
mv_k_rmse_results["five best features"]=five_best_rmse

best_six_features=["highway-mpg", "horsepower", "city-mpg", "curb-weight", "width", "length"]
six_best_rmse=knn_train_test(best_six_features, "price", cars_updated)
mv_k_rmse_results["six best features"]=six_best_rmse

    
# Output results
mv_k_rmse_results
```

Top three best models are two, three and four features. We will now further tweek them by selecting k values across a big range

```python
def knn_train_test(training_cols, target_col, df):
    np.random.seed(1)
    
    # Randomize order of rows in data frame.
    shuffled_index = np.random.permutation(df.index)
    random_df = df.reindex(shuffled_index)
    
    # Divide dataset in half for train, test
    midsplit=round(len(random_df)/2)

    # Define train and test dataset
    train_set=df.iloc[:midsplit]
    test_set=df.iloc[midsplit:]
    
    # Different values of k between 1 to 25
    k_values = [i for i in range(1,25)]
    mv_k_rmse={}
    
    for options in k_values:
        knn=KNeighborsRegressor(n_neighbors=options)

        # KNN fit and predict
        knn.fit(train_set[training_cols], train_set[target_col])
        prediction=knn.predict(test_set[training_cols])
    
        # Calculate mse and rmse
        mse=mean_squared_error(test_set[[target_col]], prediction)
        rmse=np.sqrt(mse)
        mv_k_rmse[options]=rmse
       
    return mv_k_rmse

mv_k_rmse_results={}
training_columns=cars_updated.columns.drop("price")

# Pass to function 2, 3, 4 feature set
best_two_features=["highway-mpg", "horsepower"]
two_best_rmse=knn_train_test(best_two_features, "price", cars_updated)
mv_k_rmse_results["two best features"]=two_best_rmse

best_three_features=["highway-mpg", "horsepower", "city-mpg"]
three_best_rmse=knn_train_test(best_three_features, "price", cars_updated)
mv_k_rmse_results["three best features"]=three_best_rmse

best_four_features=["highway-mpg", "horsepower", "city-mpg", "curb-weight"]
four_best_rmse=knn_train_test(best_four_features, "price", cars_updated)
mv_k_rmse_results["four best features"]=four_best_rmse

    
# Output results
mv_k_rmse_results
```

```python
for k,v in mv_k_rmse_results.items():
    x=list(v.keys())
    y=list(v.values())
    
    plt.plot(x,y)
    plt.xlabel("K-Values")
    plt.ylabel("RMSE")
```
